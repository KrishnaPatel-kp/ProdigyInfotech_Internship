# Visulazation of Data

### Distribution of Ages

```python
# Custom y-axis label
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=20, kde=True, color='blue')
plt.title('Distribution of Ages')
plt.xlabel('Age')
plt.ylabel('Count of Individuals')
plt.show()
```

###  Distribution of Genders 

```python
gender_counts = df_clean['Gender'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=gender_counts.index, y=gender_counts.values, palette='viridis')
plt.title('Distribution of Genders')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

### Distribution of Salaries

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Salary'], bins=20, kde=True, color='green')
plt.title('Distribution of Salaries')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

### Distribution of Years of Experience

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Years of Experience'], bins=20, kde=True, color='purple')
plt.title('Distribution of Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Frequency')
plt.show()
```

### Box Plot of Salary by Gender

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Gender', y='Salary', data=df, palette='Set3')
plt.title('Salary Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Salary')
plt.show()
```

### Box Plot of Years of Experience by Gender

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Gender', y='Years of Experience', data=df, palette='Set3')
plt.title('Years of Experience Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Years of Experience')
plt.show()
```

### Scatter Plot of Age vs. Salary

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Salary', hue='Gender', data=df, palette='Set1')
plt.title('Age vs. Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

### Scatter Plot of Years of Experience vs. Salary

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Years of Experience', y='Salary', hue='Gender', data=df, palette='Set1')
plt.title('Years of Experience vs. Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
